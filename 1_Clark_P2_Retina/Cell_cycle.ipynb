# Cell cycle analysis

#### Initialize the function

```r
#Load library
library(Matrix)
library(tagcloud)
library(scatterplot3d)

#Load data
scaled_expression_matrix <- readRDS("./data/scaled_expression_matrix.rds")
dat_feature <- readRDS("./data/dat_feature.rds")
umap_result <- readRDS("./data/umap_result.rds")

#Decalre a function of plot
plot_cells <- function(NAME, ANGLE){
cell_expression_colors <- smoothPalette(
    scaled_expression_matrix[rownames(subset(dat_feature, gene_short_name==NAME)),]
    , pal ="Reds")

scatterplot3d(
    x=umap_result[,1], y=umap_result[,2], z=umap_result[,3], 
    xlab="UMAP1", ylab="UMAP2", zlab="UMAP3",
    pch=16, xlim=c(-5,5), angle=ANGLE, color = cell_expression_colors,
    box=FALSE
 )
}
```
